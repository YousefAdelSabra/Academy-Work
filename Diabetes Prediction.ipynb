# Diabetes Prediction with Gaussian Naive Bayes and K-Nearest Neightbour

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("D:/AAST/Semester 8/Soft Computing/Sec/Assignments/First/Diabetespred.csv")
df.head()
```

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split

x = df.iloc[:, 0:8]
y = df.iloc[:, 8:9]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=24)

print("X Train:", x_train.shape)
print("Y Train:", y_train.shape)
print("\nX Test:", x_test.shape)
print("Y Test:", y_test.shape)
```

# Gaussian Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(x_train, y_train)
y_pred_gnb = gnb.predict(x_test)
```

# K Nearest Neighbour

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(x_train, y_train)
y_pred_knn = knn.predict(x_test)
```

```python
from sklearn.metrics import precision_score, accuracy_score, recall_score, f1_score

# Gaussian Naive Bayes
print("GNB:")
print("Accuracy:   {:.2f}".format(accuracy_score(y_test, y_pred_gnb)))
print("Precision:  {:.4f}".format(precision_score(y_test, y_pred_gnb)))
print("Recall:     {:.4f}".format(recall_score(y_test, y_pred_gnb)))
print("F1 Score:   {:.4f}".format(f1_score(y_test, y_pred_gnb)))

# KNN
print("\nKNN:")
print("Accuracy:   {:.2f}".format(accuracy_score(y_test, y_pred_knn)))
print("Precision:  {:.4f}".format(precision_score(y_test, y_pred_knn)))
print("Recall:     {:.4f}".format(recall_score(y_test, y_pred_knn)))
print("F1 Score:   {:.4f}".format(f1_score(y_test, y_pred_knn)))
```
